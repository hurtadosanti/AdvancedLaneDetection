# Camera Calibration
# Import packages

```python
import cv2
import matplotlib.pyplot as plt
from utilities import calibration
from utilities import image_utility
```

## Calibrate the camera using the images on the specify path

```python
c = calibration.CameraCalibration()
response = c.calibrate_camera('./camera_cal/calibration*.jpg')
c.serialize_calibration('camera_cal')
```

## Test the calibration

```python
original = cv2.imread('./camera_cal/calibration1.jpg')
undistort,mtx,dist = c.undistort_image(original)
# Draw the results
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(40, 20))
f.tight_layout()
ax1.imshow(cv2.cvtColor(original,cv2.COLOR_BGR2RGB))
ax1.set_title('original', fontsize=50)
ax2.imshow(cv2.cvtColor(undistort,cv2.COLOR_BGR2RGB))
ax2.set_title('undistorted', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
cv2.imwrite('results/original.jpg',cv2.resize(original,None,fx=0.3,fy=0.3))
cv2.imwrite('results/undistort.jpg',cv2.resize(undistort,None,fx=0.3,fy=0.3))
```
